```python
from random import randint
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
Ne=7
Ns=30
Glaider=np.zeros((Ne,Ne))
Glaider[1,3]=1
Glaider[2,1]=1
Glaider[2,3]=1
Glaider[3,2]=1
Glaider[3,3]=1
A=Glaider
#A=np.random.randint(2, size=(Ne, Ne))
B=np.zeros((Ne,Ne))
print(A)
```

```python
def NeibSum(i,j):
    
    return A[i-1,j-1]+A[i-1,j]+A[i-1,(j+1)%Ne]+A[i,j-1]+A[i,(j+1)%Ne]+A[(i+1)%Ne,(j-1)]+A[(i+1)%Ne,j]+A[(i+1)%Ne,(j+1)%Ne]

Game=[A]
for n in range(Ns-1):
    for i in range(0,Ne):
        for j in range(0,Ne):
            b=NeibSum(i,j)
            if b<2:
                B[i,j]=0
            elif b==2:
                B[i,j]=A[i,j]
            elif b>3:
                B[i,j]=0
            elif b==3:
                B[i,j]=1
    Game.append(B)  
    A=B  
    B=np.zeros((Ne,Ne))
```
